#Defining dependent and independent variables

```python
x=car_df.drop(['Customer Name','Country','Customer e-mail','Car Purchase Amount'],axis=1)
```

#Feature Scaling

```python
from sklearn.preprocessing import MinMaxScaler
scaler=MinMaxScaler()
x_scaled=scaler.fit_transform(x)
```

#Values are scaled between 0 to 1 and maximum and minimum values are found for the attributes

```python
scaler.data_max_
```

```python
scaler.data_min_
```

```python
y=y.values.reshape(-1,1)
```

```python
y
```

```python
y.shape
```

```python
y_scaled=scaler.fit_transform(y)
```

```python
y_scaled
```

#Training the model

```python
from sklearn.model_selection import train_test_split
```

```python
x_train,x_test,y_train,y_test=train_test_split(x_scaled,y_scaled,test_size=0.25)
```

```python
x_train.shape
```

```python
x_test.shape
```

#model training

```python
import tensorflow.keras
from keras.models import Sequential
from keras.layers import Dense
model=Sequential()
model.add(Dense(50,input_dim=5,activation='relu'))
model.add(Dense(50,activation='relu'))
model.add(Dense(1,activation='linear'))
```

```python
model.summary()
```

```python
model.compile(optimizer='adam',loss='mean_squared_error',metrics=[tensorflow.keras.metrics.RootMeanSquaredError()])
```

```python
epochs_hist=model.fit(x_train,y_train,epochs=100,batch_size=10,verbose=1,validation_split=0.2)
```

```python
  model.compile(optimizer='adam',loss='mean_squared_error',metrics=[tensorflow.keras.metrics.RootMeanSquaredError()])
```

```python
epochs_hist.history.keys()
```

```python
plt.plot(epochs_hist.history['loss'])
plt.plot(epochs_hist.history['val_loss'])
plt.title('Managing Loss')
plt.xlabel('Epoch')
plt.ylabel('Training and Validation_Loss')
plt.legend(['Training_Loss','Validation_Loss'])
```

```python
x_test=np.array([[1,45,20000,10000,340000]])
y_pred=model.predict(x_test)
```
